```python
import torch
from torch import nn, optim
import torch.nn.functional as F
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

m=[]
n=[]    
with open('cords.txt','r') as f:
  strings= f.readlines()
  for string in strings:
    string = string.split()   
    m.append([int(string[0]),int(string[1])])
    n.append([int(string[2]),int(string[3])])
print(m)
print(n)
m_t = []
n_t = []
for i in range(len(m)):
  if i%50==0:
    m_t.append(m[i])
    n_t.append(n[i])
    del m[i]
    del n[i]
print(m)
print(n)
print(m_t)
print(n_t)
```

```python
class Net(nn.Module):
    def __init__(self,neyrons):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 10)
        self.fc2 = nn.Linear(10, 100)
        self.fc3 = nn.Linear(100, 150)
        self.fc4 = nn.Linear(150, 100)
        self.fc5 = nn.Linear(100, 10)
        self.fc6 = nn.Linear(10, 2)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = F.relu(self.fc5(x))
        x = self.fc6(x)
        return x

DEVICE = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
net = Net(100)
net.to(DEVICE)
optimizer = optim.Adam(net.parameters(), lr=0.01)
net.train()


def loss(pred, target):
    squares = (pred - target) ** 2
    return squares.mean()


#x_train = [[1212,2121],[1232,21323],[213123,314134]]
#x_train = torch.Tensor(x_train)
#y_train = torch.Tensor(y_train)



x_train = torch.Tensor(m) / 3000
y_train = torch.Tensor(n) / 3000

x_train.unsqueeze_(1)
y_train.unsqueeze_(1)

x_train = x_train.to(DEVICE)
y_train = y_train.to(DEVICE)

for epoch_index in range(100):
    optimizer.zero_grad()
    y_pred = net(x_train)
    
    loss_val = loss(y_pred, y_train)
    print(loss_val.to("cpu").item())
    loss_val.backward()
    optimizer.step()



x_test = torch.Tensor([[1314, 910]]) / 3000
x_test.unsqueeze_(1)
x_test = x_test.to(DEVICE)
y_pred = net.forward(x_test) * 3000

y_pred = list(y_pred.detach().cpu().numpy())
print(y_pred)
# for i in range(len(y_pred)):
#   print(n[i],y_pred[i])





# plt.plot(x_test.detach().cpu().numpy(), y_pred.detach().cpu().numpy(), 'o')
# plt.xlabel('$x$')
# plt.ylabel('$y$')

# print("test:", m_t)
# x_test = torch.Tensor(m_t)
# x_test.unsqueeze_(1)
# x_test = x_test.to(DEVICE)
# y_pred = net.forward(x_test)

# y_pred = list(y_pred.detach().cpu().numpy())

# for i in range(len(y_pred)):
#   print(f"Raw: ({m_t[i][0], m_t[i][1]})")
#   x, y = n_t[i][0] - y_pred[i][0][0], n_t[i][1] - y_pred[i][0][1]
#   print(f"Predicted: ({x, y})")


# 990 223

'''
plt.plot(x_test.detach().cpu().numpy(), y_pred.detach().cpu().numpy(), 'o')
plt.xlabel('$x$')
plt.ylabel('$y$')
'''

torch.save(net, 'model.t')
```
